```python
# import dependencies
import pandas as pd
```

```python
# load data as dataFrame
train_dataset = pd.read_csv('/content/heartbeat/mitbih_train.csv')
# df.fillna(0, inplace=True)
train_dataset
```

```python
x = train_dataset.iloc[:,:-1]
x
```

```python
y = train_dataset.iloc[:,-1]
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

```python
# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=12)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)
```

```python
y_pred_test=knn.predict(X_test)
print('accuracy = ',accuracy_score(y_test,y_pred_test).round(2)*100)
recall = recall_score(y_test, y_pred_test, average=None)
print('recall = ',recall.round(2)*100)
```

```python
from sklearn.svm import SVC
```

```python
# initialize the model
svm_model = SVC(random_state=42)
# train the model
svm_model.fit(X_train, y_train)
```

```python
# prediction of our data
y_pred = svm_model.predict(X_test)
y_pred
```

```python
print('accuracy = ',accuracy_score(y_test, y_pred).round(2)*100)
recall = recall_score(y_test, y_pred_test, average=None)
print('recall = ',recall.round(2)*100)
```
